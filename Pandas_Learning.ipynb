## Section 5: Data Cleaning and Handling Missing Values
Handle missing values using dropna(), fillna(), and detect NULL values with isnull() and notnull().

```python
# Create DataFrame with missing values
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, 7, 8, np.nan],
    'C': [10, 20, 30, 40, 50]
})

print("DataFrame with Missing Values:")
print(df_missing)
print("\n")

# Detect missing values
print("Detect NULL values (isnull()):")
print(df_missing.isnull())
print("\n")

print("Count of NULL values per column:")
print(df_missing.isnull().sum())
print("\n")

# Drop rows with any missing values
print("After dropna() - Remove rows with ANY missing values:")
print(df_missing.dropna())
print("\n")

# Fill missing values with a constant
print("After fillna(0) - Fill NaN with 0:")
print(df_missing.fillna(0))
print("\n")

# Forward fill - propagate last valid value
print("After fillna(method='ffill') - Forward fill:")
print(df_missing.fillna(method='ffill'))
print("\n")

# Fill with column mean
print("After fillna with column mean:")
print(df_missing.fillna(df_missing.mean()))
```

Apply transformations using apply(), map(), rename columns, and create new columns through calculations.

## Section 7: Grouping and Aggregation
Use groupby() to group data and aggregate with sum(), mean(), count(), and custom aggregation functions.

```python
# Create sample data for grouping
sales_data = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'East', 'West', 'South', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Sales': [100, 150, 200, 175, 125, 100, 250, 180],
    'Quantity': [10, 15, 20, 18, 13, 10, 25, 18]
})

print("Sample Data:")
print(sales_data)
print("\n")

# Group by single column and calculate mean
print("Average Sales by Region:")
print(sales_data.groupby('Region')['Sales'].mean())
print("\n")

# Group by single column and calculate multiple aggregations
print("Sales statistics by Region:")
print(sales_data.groupby('Region')['Sales'].agg(['sum', 'mean', 'count', 'min', 'max']))
print("\n")

# Group by multiple columns
print("Sales by Region and Product:")
print(sales_data.groupby(['Region', 'Product'])['Sales'].sum())
print("\n")

# Custom aggregation function
def salary_range(x):
    return x.max() - x.min()

print("Sales range by Region:")
print(sales_data.groupby('Region')['Sales'].agg(salary_range))
print("\n")

# Multiple aggregations on multiple columns
print("Multiple aggregations:")
agg_result = sales_data.groupby('Region').agg({
    'Sales': ['sum', 'mean'],
    'Quantity': ['sum', 'mean']
})
print(agg_result)
```

## Section 8: Merging and Joining DataFrames
Combine DataFrames using merge(), join(), and concat() with different join types.

```python
# Create two DataFrames to merge
df_employees = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Department': ['Sales', 'IT', 'HR', 'Finance']
})

df_salaries = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 5],  # Note: 5 is not in employees, will show how joins work
    'Salary': [50000, 60000, 55000, 70000]
})

print("Employees DataFrame:")
print(df_employees)
print("\nSalaries DataFrame:")
print(df_salaries)
print("\n")

# Inner Join (only matching rows)
print("INNER JOIN (only matching EmployeeIDs):")
inner_join = pd.merge(df_employees, df_salaries, on='EmployeeID', how='inner')
print(inner_join)
print("\n")

# Left Join (all from left, matching from right)
print("LEFT JOIN (all from left):")
left_join = pd.merge(df_employees, df_salaries, on='EmployeeID', how='left')
print(left_join)
print("\n")

# Right Join (all from right, matching from left)
print("RIGHT JOIN (all from right):")
right_join = pd.merge(df_employees, df_salaries, on='EmployeeID', how='right')
print(right_join)
print("\n")

# Outer Join (all rows from both)
print("OUTER JOIN (all rows):")
outer_join = pd.merge(df_employees, df_salaries, on='EmployeeID', how='outer')
print(outer_join)
print("\n")

# Concatenate DataFrames (row-wise)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
print("Concatenate DataFrames:")
print(pd.concat([df1, df2], ignore_index=True))
```

## Section 11: Pivot Tables & Cross-tabulations
Reshape data for analysis using pivot operations.

```python
# Pivot Tables & Cross-tabulations

pivot_data = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'] * 2,
    'Region': ['North', 'South', 'North', 'South', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
    'Sales': [100, 150, 200, 120, 180, 140, 160, 190]
})

print("Original Data:")
print(pivot_data.head(8))
print("\n")

# Create pivot table
pivot = pd.pivot_table(
    pivot_data,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)
print("Pivot Table (sum by Region and Product):")
print(pivot)
print("\n")

# Pivot table with multiple aggregations
pivot_agg = pd.pivot_table(
    pivot_data,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc=['sum', 'mean', 'count']
)
print("Pivot Table with Multiple Aggregations:")
print(pivot_agg)
print("\n")

# Crosstab - cross-tabulation
ct = pd.crosstab(pivot_data['Region'], pivot_data['Product'], values=pivot_data['Sales'], aggfunc='sum')
print("Crosstab (Region vs Product, summed Sales):")
print(ct)
print("\n")

# Crosstab with margins (totals)
ct_margins = pd.crosstab(pivot_data['Region'], pivot_data['Product'], margins=True)
print("Crosstab with Margins (counts):")
print(ct_margins)
```

## Section 12: Time Series & DateTime
Handle dates, time-based operations, and resampling.

```python
# Time Series & DateTime Operations

# Create date range
dates = pd.date_range('2024-01-01', periods=30, freq='D')
print("Date range (30 days):")
print(dates[:5], "...", dates[-1])
print("\n")

# Create time series data
ts_data = pd.DataFrame({
    'Date': dates,
    'Value': np.random.randint(50, 200, 30),
    'Price': np.random.uniform(10, 50, 30)
})

print("Time Series Data (first 5 rows):")
print(ts_data.head())
print("\n")

# Set date as index
ts_data_indexed = ts_data.set_index('Date')
print("With Date as Index:")
print(ts_data_indexed.head())
print("\n")

# Resampling - Resample to weekly (mean)
weekly = ts_data_indexed.resample('W').mean()
print("Weekly Average (Resampled):")
print(weekly)
print("\n")

# Resampling - Different aggregations
print("Weekly - Last value of week:")
print(ts_data_indexed['Value'].resample('W').last())
print("\n")

# Date arithmetic
ts_data['Date'] = pd.to_datetime(ts_data['Date'])
ts_data['Month'] = ts_data['Date'].dt.month
ts_data['DayOfWeek'] = ts_data['Date'].dt.day_name()
ts_data['Quarter'] = ts_data['Date'].dt.quarter

print("After extracting date components:")
print(ts_data[['Date', 'Month', 'DayOfWeek', 'Quarter']].head())
print("\n")

# Filter time series data
print("Data from 2024-01-15 to 2024-01-25:")
mask = (ts_data['Date'] >= '2024-01-15') & (ts_data['Date'] <= '2024-01-25')
print(ts_data[mask].head())
```

## Section 13: Window Functions
Apply rolling, expanding, and exponential functions for time series analysis.

```python
# Window Functions for Time Series Analysis

# Create sample stock price data
stock_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=20),
    'Price': [100, 102, 101, 105, 103, 108, 107, 110, 109, 112,
              111, 115, 113, 116, 114, 118, 117, 120, 119, 122]
})

print("Original Stock Prices:")
print(stock_data)
print("\n")

# Rolling average (moving average)
stock_data['MA_3'] = stock_data['Price'].rolling(window=3).mean()
stock_data['MA_5'] = stock_data['Price'].rolling(window=5).mean()
print("Moving Averages (3-day and 5-day):")
print(stock_data[['Date', 'Price', 'MA_3', 'MA_5']].head(10))
print("\n")

# Rolling standard deviation
stock_data['Std_Dev_3'] = stock_data['Price'].rolling(window=3).std()
print("Rolling Standard Deviation (3-day):")
print(stock_data[['Date', 'Price', 'Std_Dev_3']].head())
print("\n")

# Expanding window (grows as you move forward)
stock_data['Expanding_Mean'] = stock_data['Price'].expanding().mean()
stock_data['Expanding_Max'] = stock_data['Price'].expanding().max()
print("Expanding Mean and Max:")
print(stock_data[['Date', 'Price', 'Expanding_Mean', 'Expanding_Max']].head(8))
print("\n")

# Exponential moving average (more weight to recent values)
stock_data['EMA'] = stock_data['Price'].ewm(span=5).mean()
print("Exponential Moving Average:")
print(stock_data[['Date', 'Price', 'EMA']].head(10))
print("\n")

# Daily returns (percentage change)
stock_data['Daily_Return'] = stock_data['Price'].pct_change()
print("Daily Returns (percentage change):")
print(stock_data[['Date', 'Price', 'Daily_Return']].head())
```

## Section 14: Categorical Data
Work with categorical data for memory optimization and specialized operations.

```python
# Categorical Data

# Create DataFrame with string categories
df_cat = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Size': ['Small', 'Medium', 'Large', 'Medium'],
    'Priority': ['High', 'Low', 'High', 'Medium']
})

print("Original DataFrame (string type):")
print(df_cat.dtypes)
print("\n")

# Convert to categorical
df_cat['Size'] = pd.Categorical(df_cat['Size'], categories=['Small', 'Medium', 'Large'], ordered=True)
df_cat['Priority'] = pd.Categorical(df_cat['Priority'], categories=['Low', 'Medium', 'High'], ordered=True)

print("After converting to Categorical:")
print(df_cat.dtypes)
print("\n")

# Benefits: Less memory usage
print("Memory usage (strings):", df.memory_usage(deep=True).sum(), "bytes")
print("Memory usage (categorical):", df_cat.memory_usage(deep=True).sum(), "bytes")
print("\n")

# Value counts on categorical
print("Priority value counts:")
print(df_cat['Priority'].value_counts())
print("\n")

# Ordered categorical comparisons
print("Sizes >= 'Medium':")
print(df_cat[df_cat['Size'] >= 'Medium'])
print("\n")

# Rename categories
df_cat['Priority'] = df_cat['Priority'].cat.rename_categories({'High': 'Critical', 'Low': 'Minor'})
print("After renaming categories:")
print(df_cat['Priority'].cat.categories)
print("\n")

# Add new category
df_cat['Priority'] = df_cat['Priority'].cat.add_categories(['Urgent'])
print("After adding 'Urgent' category:")
print(df_cat['Priority'].cat.categories)
```

## Section 15: Advanced Reshaping
Reshape data using melt, transpose, and other reshaping techniques.

```python
# Advanced Reshaping

# Wide format data (each observation in columns)
wide_data = pd.DataFrame({
    'ID': [1, 2, 3],
    'Q1': [100, 150, 120],
    'Q2': [110, 160, 130],
    'Q3': [120, 170, 140],
    'Q4': [130, 180, 150]
})

print("Wide Format Data:")
print(wide_data)
print("\n")

# Melt - Convert wide to long format
long_data = pd.melt(wide_data, id_vars=['ID'], var_name='Quarter', value_name='Sales')
print("Long Format Data (melted):")
print(long_data)
print("\n")

# Pivot (reverse of melt) - Convert long to wide
pivoted = long_data.pivot(index='ID', columns='Quarter', values='Sales')
print("Back to Wide Format (pivoted):")
print(pivoted)
print("\n")

# Transpose - Switch rows and columns
transposed = wide_data.set_index('ID').T
print("Transposed DataFrame:")
print(transposed)
print("\n")

# Explode - Convert list values in cells to rows
df_list = pd.DataFrame({
    'ID': [1, 2, 3],
    'Tags': [['A', 'B'], ['C', 'D', 'E'], ['F']]
})

print("DataFrame with lists:")
print(df_list)
print("\n")

exploded = df_list.explode('Tags')
print("After explode (one tag per row):")
print(exploded)
print("\n")

# Get dummies - One-hot encoding
df_dummies = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Red', 'Green']
})
dummies = pd.get_dummies(df_dummies['Color'])
print("One-hot encoding (get_dummies):")
print(dummies)
```

## Section 16: File I/O Operations
Read and write DataFrames in various file formats.

```python
# File I/O Operations

import os

# Create sample data
io_data = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Salary': [50000, 60000, 75000, 55000, 70000]
})

# Create directory for saving files
save_dir = "pandas_files"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Save to CSV
csv_path = os.path.join(save_dir, "data.csv")
io_data.to_csv(csv_path, index=False)
print(f"Saved to CSV: {csv_path}")

# Read from CSV
df_csv = pd.read_csv(csv_path)
print("Read from CSV:")
print(df_csv.head())
print("\n")

# Save to Excel (if openpyxl is installed)
try:
    excel_path = os.path.join(save_dir, "data.xlsx")
    io_data.to_excel(excel_path, index=False, sheet_name='Data')
    print(f"Saved to Excel: {excel_path}")
except:
    print("Excel support requires 'openpyxl' package")
print("\n")

# Save to JSON
json_path = os.path.join(save_dir, "data.json")
io_data.to_json(json_path, orient='records', indent=2)
print(f"Saved to JSON: {json_path}")

# Read from JSON
df_json = pd.read_json(json_path)
print("Read from JSON:")
print(df_json.head())
print("\n")

# Save to Parquet (if pyarrow/fastparquet installed)
try:
    parquet_path = os.path.join(save_dir, "data.parquet")
    io_data.to_parquet(parquet_path)
    print(f"Saved to Parquet: {parquet_path}")
except:
    print("Parquet support requires 'pyarrow' or 'fastparquet'")
print("\n")

# Save to HDF5 (if tables installed)
try:
    hdf5_path = os.path.join(save_dir, "data.h5")
    io_data.to_hdf(hdf5_path, key='data')
    print(f"Saved to HDF5: {hdf5_path}")
except:
    print("HDF5 support requires 'tables' package")

print("\nFile I/O operations completed!")
```

## Section 17: Performance Optimization
Optimize pandas operations for large datasets.

```python
# Performance Optimization

import time

# Create large dataset
large_df = pd.DataFrame({
    'A': np.random.randint(0, 100, 100000),
    'B': np.random.randint(0, 100, 100000),
    'C': np.random.choice(['X', 'Y', 'Z'], 100000)
})

print("Large DataFrame shape:", large_df.shape)
print("Memory usage:", large_df.memory_usage(deep=True).sum(), "bytes")
print("\n")

# Method 1: Slow - Using apply with lambda
start = time.time()
large_df['D_slow'] = large_df.apply(lambda row: row['A'] + row['B'], axis=1)
slow_time = time.time() - start
print(f"Apply method time: {slow_time:.4f} seconds")

# Method 2: Fast - Vectorized operation
start = time.time()
large_df['D_fast'] = large_df['A'] + large_df['B']
fast_time = time.time() - start
print(f"Vectorized method time: {fast_time:.4f} seconds")
print(f"Speedup: {slow_time / fast_time:.1f}x faster")
print("\n")

# Optimize dtypes to save memory
print("Original dtypes:")
print(large_df.dtypes)
print("Memory usage:", large_df.memory_usage(deep=True).sum(), "bytes")
print("\n")

# Downcast integers
large_df['A'] = pd.to_numeric(large_df['A'], downcast='integer')
# Convert string to category
large_df['C'] = large_df['C'].astype('category')

print("Optimized dtypes:")
print(large_df.dtypes)
print("Memory usage after optimization:", large_df.memory_usage(deep=True).sum(), "bytes")
print("\n")

# Use query() for filtering (faster than boolean indexing for complex conditions)
start = time.time()
result1 = large_df[(large_df['A'] > 50) & (large_df['B'] < 30)]
bool_time = time.time() - start

start = time.time()
result2 = large_df.query('A > 50 and B < 30')
query_time = time.time() - start

print(f"Boolean indexing time: {bool_time:.4f} seconds")
print(f"Query method time: {query_time:.4f} seconds")
print("\n")

# Chunking - Process large files in batches
print("Reading file in chunks (example):")
chunk_size = 10000
chunks = pd.read_csv("pandas_files/data.csv", chunksize=chunk_size, nrows=30000) if os.path.exists("pandas_files/data.csv") else None
if chunks:
    for i, chunk in enumerate(chunks):
        print(f"  Chunk {i}: {len(chunk)} rows")
        if i >= 2:  # Just show first 3 chunks
            break
```

```python
# String Operations

str_data = pd.DataFrame({
    'Text': ['Hello World', 'PYTHON PANDAS', 'data Analysis', 'MACHINE learning'],
    'Email': ['alice@example.com', 'bob_smith@test.org', 'charlie123@mail.co.uk', 'diana.jones@work.net']
})

print("Original Data:")
print(str_data)
print("\n")

# Basic string operations
print("Uppercase:")
print(str_data['Text'].str.upper())
print("\n")

print("Lowercase:")
print(str_data['Text'].str.lower())
print("\n")

print("Length:")
print(str_data['Text'].str.len())
print("\n")

# String slicing
print("First 5 characters:")
print(str_data['Text'].str[:5])
print("\n")

# Replace
print("Replace 'a' with '*':")
print(str_data['Text'].str.replace('a', '*', case=False))
print("\n")

# Split
print("Split by space:")
print(str_data['Text'].str.split(' '))
print("\n")

# Extract domain from email
print("Extract domain from email:")
print(str_data['Email'].str.extract(r'@(.+)', expand=False))
print("\n")

# Contains check
print("Texts containing 'a' (case-insensitive):")
print(str_data['Text'].str.contains('a', case=False))
print("\n")

# Strip whitespace
str_data2 = pd.DataFrame({
    'Names': ['  Alice  ', ' Bob ', 'Charlie']
})
print("Original with spaces:")
print(repr(str_data2['Names']))
print("\nAfter strip:")
print(repr(str_data2['Names'].str.strip()))
print("\n")

# Find index of substring
print("Index of first 'a' (case-insensitive):")
print(str_data['Text'].str.find('a'))
print("\n")

# Count occurrences
print("Count of 'a' in each text:")
print(str_data['Text'].str.count('a', case=False))
```

## Section 19: Data Validation & Type Conversion
Handle type conversion, validate data quality, and manage data types.

```python
# Data Validation & Type Conversion

val_data = pd.DataFrame({
    'ID': ['1', '2', 'invalid', '4', '5'],
    'Price': ['100.5', '200', 'N/A', '150.75', '220'],
    'Quantity': ['10', '20', '30', 'undefined', '50'],
    'Active': ['true', 'false', 'true', 'maybe', 'false']
})

print("Original Data with Mixed Types:")
print(val_data)
print(val_data.dtypes)
print("\n")

# Type conversion with astype()
print("Convert ID to numeric (will fail on 'invalid'):")
try:
    val_data['ID'] = val_data['ID'].astype('int')
except ValueError as e:
    print(f"Error: {e}")
print("\n")

# Convert with errors='coerce' - convert invalid to NaN
print("Convert ID to numeric (invalid -> NaN):")
val_data['ID'] = pd.to_numeric(val_data['ID'], errors='coerce')
print(val_data['ID'])
print("\n")

# Convert Price
print("Convert Price to numeric (N/A -> NaN):")
val_data['Price'] = pd.to_numeric(val_data['Price'], errors='coerce')
print(val_data['Price'])
print("\n")

# Convert Quantity
print("Convert Quantity to numeric (undefined -> NaN):")
val_data['Quantity'] = pd.to_numeric(val_data['Quantity'], errors='coerce')
print(val_data['Quantity'])
print("\n")

# Convert boolean
print("Convert Active to boolean:")
val_data['Active'] = val_data['Active'].map({'true': True, 'false': False}, na_action='ignore')
print(val_data['Active'])
print("\n")

# Check for data quality issues
print("Data Quality Report:")
print(f"Null values per column:")
print(val_data.isnull().sum())
print("\n")

print("Percentage of missing data per column:")
print((val_data.isnull().sum() / len(val_data) * 100).round(2))
print("\n")

# Validate ranges
print("Rows where Price > 500:")
invalid_prices = val_data[val_data['Price'] > 500]
print(invalid_prices if len(invalid_prices) > 0 else "None")
print("\n")

# Validate duplicates
val_data2 = pd.DataFrame({
    'User': ['Alice', 'Bob', 'Alice', 'Charlie'],
    'Score': [100, 95, 100, 88]
})
print("Check for duplicates:")
print(val_data2[val_data2.duplicated(subset=['User'], keep=False)])
print("\n")

# Drop duplicates
print("After dropping duplicates (keeping first):")
print(val_data2.drop_duplicates(subset=['User'], keep='first'))
print("\n")

# Using convert_dtypes() for better type inference
print("Using convert_dtypes() for automatic type detection:")
val_data_converted = val_data.convert_dtypes(infer_objects=True)
print(val_data_converted.dtypes)
```

## Section 20: Advanced Tips & Best Practices
Bonus section with method chaining and other professional practices.

```python
# Advanced Tips & Best Practices

# Method Chaining - Write clean, readable pipelines
data_chain = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Sales': [100, 150, 200, 120, 160, 180],
    'Region': ['North', 'South', 'East', 'North', 'South', 'East']
})

print("Original Data:")
print(data_chain)
print("\n")

# Instead of nested operations, use method chaining for clarity
result = (data_chain
    .query("Sales > 110")  # Filter
    .groupby('Product')['Sales'].sum()  # Group and aggregate
    .sort_values(ascending=False)  # Sort
)

print("Method Chaining Result (sells > 110, grouped, sorted):")
print(result)
print("\n")

# Use assign() for adding columns with method chaining
result2 = (data_chain
    .assign(Sales_Category=lambda x: pd.cut(x['Sales'], bins=[0, 140, 200], labels=['Low', 'High']))
    .assign(Sales_Doubled=lambda x: x['Sales'] * 2)
)

print("Using assign() for method chaining:")
print(result2)
print("\n")

# Use pipe() for custom functions in chains
def filter_and_aggregate(df):
    return df.groupby('Product')['Sales'].agg(['sum', 'mean'])

result3 = data_chain.pipe(filter_and_aggregate)
print("Using pipe() for custom functions:")
print(result3)
print("\n")

# Copy DataFrame to avoid SettingWithCopyWarning
original = pd.DataFrame({'A': [1, 2, 3]})
modified = original.copy()  # Always copy when modifying subsets
modified['B'] = [4, 5, 6]
print("Using .copy() to avoid side effects:")
print("Original shape:", original.shape)
print("Modified shape:", modified.shape)
print("\n")

# Context settings for display
print("Display settings example:")
with pd.option_context('display.max_rows', 3, 'display.max_columns', 1):
    print("Limited display:")
    print(data_chain)

print("\n=== Congratulations! You've completed Pandas Learning! ===")
print("\nKey Takeaways:")
print("✓ DataFrames and Series are the core data structures")
print("✓ Use vectorized operations instead of loops for performance")
print("✓ Method chaining makes code more readable and maintainable")
print("✓ Always handle missing data explicitly")
print("✓ Use appropriate data types to optimize memory and speed")
print("✓ Groupby and aggregation are essential for data analysis")
print("✓ Time series analysis requires understanding of datetime operations")
print("\nNext Steps: Apply these skills to real-world datasets!")
print("Happy Data Wrangling! 📊💻")
```
