## 9. Advanced Parameterized Calls

Handle complex parameterized requests with multiple filters, dynamic queries, and flexible parameter combinations.

### Types of Parameterized Calls:
- **Multiple Query Filters**: Combine multiple parameters for advanced filtering
- **Optional Parameters**: Make parameters optional with default values
- **Parameter Validation**: Validate and constrain parameter values
- **Dynamic Query Building**: Build queries based on provided parameters

````python
# Simulated database
fake_products = [
    {"id": 1, "name": "Laptop", "category": "electronics", "price": 999.99, "stock": 10},
    {"id": 2, "name": "Mouse", "category": "electronics", "price": 25.50, "stock": 50},
    {"id": 3, "name": "Desk", "category": "furniture", "price": 299.99, "stock": 5},
    {"id": 4, "name": "Chair", "category": "furniture", "price": 199.99, "stock": 15},
    {"id": 5, "name": "Monitor", "category": "electronics", "price": 349.99, "stock": 8},
]

# Advanced parameterized query with multiple filters
@app.get("/products-advanced")
async def get_products_advanced(
    category: Optional[str] = Query(None, description="Filter by category"),
    min_price: Optional[float] = Query(None, gt=0, description="Minimum price"),
    max_price: Optional[float] = Query(None, gt=0, description="Maximum price"),
    in_stock: Optional[bool] = Query(None, description="Only items in stock"),
    skip: int = Query(0, ge=0, description="Skip items"),
    limit: int = Query(10, ge=1, le=100, description="Limit results"),
):
    """
    Advanced product search with multiple parameterized filters.
    All parameters are optional for flexible querying.
    """
    results = fake_products
    
    # Filter by category
    if category:
        results = [p for p in results if p["category"].lower() == category.lower()]
    
    # Filter by price range
    if min_price:
        results = [p for p in results if p["price"] >= min_price]
    if max_price:
        results = [p for p in results if p["price"] <= max_price]
    
    # Filter by stock availability
    if in_stock:
        results = [p for p in results if p["stock"] > 0]
    
    # Apply pagination
    results = results[skip:skip + limit]
    
    return {
        "total": len(fake_products),
        "filters_applied": {
            "category": category,
            "price_range": (min_price, max_price),
            "in_stock": in_stock
        },
        "items": results
    }

# Parameterized search with sorting
@app.get("/search-sorted")
async def search_with_sorting(
    query: str = Query(..., min_length=1, description="Search term"),
    sort_by: str = Query("name", regex="^(name|price|stock)$", description="Sort field"),
    sort_order: str = Query("asc", regex="^(asc|desc)$", description="Sort order"),
):
    """
    Search products with parameterized sorting options.
    """
    # Filter products by name or category
    results = [
        p for p in fake_products 
        if query.lower() in p["name"].lower() or query.lower() in p["category"].lower()
    ]
    
    # Apply sorting based on parameters
    reverse = sort_order.lower() == "desc"
    results = sorted(results, key=lambda x: x.get(sort_by, 0), reverse=reverse)
    
    return {
        "query": query,
        "sort": {"by": sort_by, "order": sort_order},
        "results": results,
        "count": len(results)
    }

# Dynamic filtering with custom parameters
class FilterParams(BaseModel):
    fields: List[str] = ["id", "name", "price"]  # Specify which fields to return
    filters: Optional[dict] = None  # Custom filter conditions
    sort: Optional[str] = None

@app.post("/products-dynamic")
async def get_products_dynamic(params: FilterParams):
    """
    Dynamic product retrieval with parameterized field selection and filtering.
    """
    results = fake_products
    
    # Apply custom filters if provided
    if params.filters:
        for key, value in params.filters.items():
            if key in ["category"]:
                results = [p for p in results if p.get(key) == value]
            elif key in ["price"]:
                if isinstance(value, dict):
                    if "min" in value:
                        results = [p for p in results if p["price"] >= value["min"]]
                    if "max" in value:
                        results = [p for p in results if p["price"] <= value["max"]]
    
    # Select only requested fields
    filtered_results = []
    for product in results:
        filtered_product = {k: v for k, v in product.items() if k in params.fields}
        filtered_results.append(filtered_product)
    
    return {
        "requested_fields": params.fields,
        "filters_used": params.filters,
        "results": filtered_results
    }

# Parameterized pagination
@app.get("/paginated/{resource_type}")
async def get_paginated(
    resource_type: str,
    page: int = Query(1, ge=1, description="Page number"),
    page_size: int = Query(5, ge=1, le=50, description="Items per page"),
):
    """
    Generic parameterized pagination endpoint.
    """
    if resource_type.lower() == "products":
        items = fake_products
    else:
        raise HTTPException(status_code=404, detail=f"Resource type '{resource_type}' not found")
    
    total_items = len(items)
    total_pages = (total_items + page_size - 1) // page_size
    
    # Validate page number
    if page > total_pages and total_items > 0:
        raise HTTPException(
            status_code=400,
            detail=f"Page {page} exceeds total pages ({total_pages})"
        )
    
    start_idx = (page - 1) * page_size
    end_idx = start_idx + page_size
    paginated_items = items[start_idx:end_idx]
    
    return {
        "resource": resource_type,
        "pagination": {
            "page": page,
            "page_size": page_size,
            "total_items": total_items,
            "total_pages": total_pages
        },
        "items": paginated_items
    }

print("✓ Advanced parameterized call handlers defined!")
````

## 10. Making API Calls and Database Operations

Fetch data from external APIs, handle responses, and store/retrieve records from a database.

### Key Topics:
- **Making HTTP Requests**: Using `httpx` or `requests` to call external APIs
- **Response Handling**: Processing API responses with error handling
- **Database Setup**: Using SQLite for simple persistence
- **CRUD Operations**: Create, Read, Update, Delete database operations
- **Async Operations**: Async/await for better performance

````python
# First, install required packages:
# pip install httpx aiosqlite sqlalchemy

import sqlite3
import json
from datetime import datetime
from contextlib import contextmanager

# ============================================================================
# DATABASE SETUP
# ============================================================================

# Create SQLite database for storing API responses
DATABASE_FILE = "api_cache.db"

def init_database():
    """Initialize database with required tables"""
    conn = sqlite3.connect(DATABASE_FILE)
    cursor = conn.cursor()
    
    # Create table for storing API responses
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS api_responses (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            source_api TEXT NOT NULL,
            endpoint TEXT NOT NULL,
            response_data TEXT NOT NULL,
            status_code INTEGER,
            fetched_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
            expires_at TIMESTAMP,
            UNIQUE(source_api, endpoint)
        )
    ''')
    
    # Create table for user-fetched data
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS fetched_data (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER,
            data_type TEXT,
            data_content TEXT,
            created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
            updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        )
    ''')
    
    conn.commit()
    conn.close()
    print("✓ Database initialized successfully!")

# Context manager for database connections
@contextmanager
def get_db_connection():
    """Manage database connections"""
    conn = sqlite3.connect(DATABASE_FILE)
    conn.row_factory = sqlite3.Row  # Returns rows as dictionaries
    try:
        yield conn
    finally:
        conn.close()

# Initialize the database
init_database()

# ============================================================================
# EXTERNAL API CALL FUNCTIONS
# ============================================================================

# Simulated external API response (for demo purposes)
def simulate_external_api_call(endpoint: str) -> dict:
    """Simulate calling an external API"""
    # In production, use httpx for async HTTP calls
    api_responses = {
        "/users": [
            {"id": 1, "name": "Alice Johnson", "email": "alice@example.com"},
            {"id": 2, "name": "Bob Smith", "email": "bob@example.com"}
        ],
        "/posts": [
            {"id": 1, "title": "FastAPI Tutorial", "content": "Learn FastAPI basics"},
            {"id": 2, "title": "Database Design", "content": "SQL best practices"}
        ]
    }
    return api_responses.get(endpoint, [])

# Store API response in database
def store_api_response(source_api: str, endpoint: str, response_data: dict, status_code: int = 200):
    """Store API response in database for caching"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        try:
            cursor.execute('''
                INSERT OR REPLACE INTO api_responses 
                (source_api, endpoint, response_data, status_code, expires_at)
                VALUES (?, ?, ?, ?, datetime('now', '+1 hour'))
            ''', (source_api, endpoint, json.dumps(response_data), status_code))
            
            conn.commit()
            print(f"✓ Stored {source_api}{endpoint} response")
            return cursor.lastrowid
        except Exception as e:
            print(f"✗ Error storing API response: {e}")
            return None

# Retrieve cached API response
def get_cached_api_response(source_api: str, endpoint: str):
    """Retrieve cached API response from database"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        cursor.execute('''
            SELECT response_data, status_code, fetched_at
            FROM api_responses
            WHERE source_api = ? AND endpoint = ? AND expires_at > datetime('now')
        ''', (source_api, endpoint))
        
        row = cursor.fetchone()
        if row:
            return {
                "data": json.loads(row['response_data']),
                "status_code": row['status_code'],
                "fetched_at": row['fetched_at'],
                "cached": True
            }
        return None

# ============================================================================
# CRUD OPERATIONS FOR USER DATA
# ============================================================================

def create_fetched_record(user_id: int, data_type: str, data_content: dict):
    """Create a new record with fetched data"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        cursor.execute('''
            INSERT INTO fetched_data (user_id, data_type, data_content)
            VALUES (?, ?, ?)
        ''', (user_id, data_type, json.dumps(data_content)))
        
        conn.commit()
        return cursor.lastrowid

def get_fetched_record(record_id: int):
    """Retrieve a stored record by ID"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        cursor.execute('''
            SELECT * FROM fetched_data WHERE id = ?
        ''', (record_id,))
        
        row = cursor.fetchone()
        if row:
            return {
                "id": row['id'],
                "user_id": row['user_id'],
                "data_type": row['data_type'],
                "data_content": json.loads(row['data_content']),
                "created_at": row['created_at'],
                "updated_at": row['updated_at']
            }
        return None

def get_user_records(user_id: int, data_type: Optional[str] = None):
    """Retrieve all records for a user, optionally filtered by type"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        if data_type:
            cursor.execute('''
                SELECT * FROM fetched_data 
                WHERE user_id = ? AND data_type = ?
                ORDER BY created_at DESC
            ''', (user_id, data_type))
        else:
            cursor.execute('''
                SELECT * FROM fetched_data 
                WHERE user_id = ?
                ORDER BY created_at DESC
            ''', (user_id,))
        
        rows = cursor.fetchall()
        return [
            {
                "id": row['id'],
                "user_id": row['user_id'],
                "data_type": row['data_type'],
                "data_content": json.loads(row['data_content']),
                "created_at": row['created_at'],
                "updated_at": row['updated_at']
            }
            for row in rows
        ]

def update_fetched_record(record_id: int, data_content: dict):
    """Update an existing record"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        cursor.execute('''
            UPDATE fetched_data 
            SET data_content = ?, updated_at = CURRENT_TIMESTAMP
            WHERE id = ?
        ''', (json.dumps(data_content), record_id))
        
        conn.commit()
        return cursor.rowcount > 0

def delete_fetched_record(record_id: int):
    """Delete a record"""
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        cursor.execute('DELETE FROM fetched_data WHERE id = ?', (record_id,))
        
        conn.commit()
        return cursor.rowcount > 0

print("✓ Database functions defined successfully!")
````

````python
# ============================================================================
# FASTAPI ENDPOINTS FOR API CALLS AND DATABASE OPERATIONS
# ============================================================================

# Endpoint to fetch from external API and cache the response
@app.get("/fetch-external/{endpoint}")
async def fetch_external_data(endpoint: str):
    """
    Fetch data from external API and cache the response in database.
    First checks cache, if expired or not found, fetches fresh data.
    """
    source_api = "https://jsonplaceholder.typicode.com"
    valid_endpoints = ["/users", "/posts"]
    
    if endpoint not in valid_endpoints:
        raise HTTPException(
            status_code=400,
            detail=f"Invalid endpoint. Valid endpoints: {valid_endpoints}"
        )
    
    # Check cache first
    cached = get_cached_api_response(source_api, endpoint)
    if cached:
        return {
            "source": source_api + endpoint,
            "data": cached["data"],
            "status": "cached",
            "fetched_at": cached["fetched_at"]
        }
    
    # If not in cache or expired, fetch fresh data
    try:
        # Simulate external API call
        api_data = simulate_external_api_call(endpoint)
        
        # Store in database
        stored_id = store_api_response(source_api, endpoint, api_data, status_code=200)
        
        return {
            "source": source_api + endpoint,
            "data": api_data,
            "status": "fresh",
            "cached_id": stored_id,
            "message": "Data fetched and cached successfully"
        }
    except Exception as e:
        raise HTTPException(
            status_code=500,
            detail=f"Failed to fetch from external API: {str(e)}"
        )

# Endpoint to store fetched data for a user
@app.post("/store-data/{user_id}")
async def store_user_data(user_id: int, data_type: str, data: dict):
    """
    Store fetched data in database associated with a user.
    """
    try:
        record_id = create_fetched_record(user_id, data_type, data)
        
        return {
            "record_id": record_id,
            "user_id": user_id,
            "data_type": data_type,
            "message": "Data stored successfully"
        }
    except Exception as e:
        raise HTTPException(
            status_code=500,
            detail=f"Failed to store data: {str(e)}"
        )

# Endpoint to retrieve a specific record
@app.get("/record/{record_id}")
async def retrieve_record(record_id: int):
    """
    Retrieve a specific stored record by ID.
    """
    record = get_fetched_record(record_id)
    
    if not record:
        raise HTTPException(
            status_code=404,
            detail=f"Record {record_id} not found"
        )
    
    return {
        "record": record,
        "message": "Record retrieved successfully"
    }

# Endpoint to retrieve all records for a user
@app.get("/user/{user_id}/records")
async def get_user_data(
    user_id: int,
    data_type: Optional[str] = Query(None, description="Filter by data type")
):
    """
    Retrieve all records for a user, optionally filtered by data type.
    """
    records = get_user_records(user_id, data_type)
    
    return {
        "user_id": user_id,
        "data_type_filter": data_type,
        "total_records": len(records),
        "records": records
    }

# Endpoint to update a record
@app.put("/record/{record_id}")
async def update_record(record_id: int, new_data: dict):
    """
    Update an existing record with new data.
    """
    record_exists = get_fetched_record(record_id)
    if not record_exists:
        raise HTTPException(
            status_code=404,
            detail=f"Record {record_id} not found"
        )
    
    success = update_fetched_record(record_id, new_data)
    
    if success:
        return {
            "record_id": record_id,
            "updated_data": new_data,
            "message": "Record updated successfully"
        }
    else:
        raise HTTPException(
            status_code=500,
            detail="Failed to update record"
        )

# Endpoint to delete a record
@app.delete("/record/{record_id}")
async def remove_record(record_id: int):
    """
    Delete a record from the database.
    """
    record_exists = get_fetched_record(record_id)
    if not record_exists:
        raise HTTPException(
            status_code=404,
            detail=f"Record {record_id} not found"
        )
    
    success = delete_fetched_record(record_id)
    
    if success:
        return {
            "record_id": record_id,
            "message": "Record deleted successfully"
        }
    else:
        raise HTTPException(
            status_code=500,
            detail="Failed to delete record"
        )

# Endpoint to fetch and store in one call
@app.post("/fetch-and-store/{user_id}/{endpoint}")
async def fetch_and_store(user_id: int, endpoint: str):
    """
    Fetch data from external API and automatically store for the user.
    Combined operation for convenience.
    """
    valid_endpoints = ["/users", "/posts"]
    
    if endpoint not in valid_endpoints:
        raise HTTPException(
            status_code=400,
            detail=f"Invalid endpoint. Valid endpoints: {valid_endpoints}"
        )
    
    try:
        # Fetch data
        api_data = simulate_external_api_call(endpoint)
        
        # Store in cache
        source_api = "https://jsonplaceholder.typicode.com"
        store_api_response(source_api, endpoint, api_data)
        
        # Store for user
        record_id = create_fetched_record(
            user_id, 
            data_type=endpoint.strip('/'),
            data_content={"source": source_api + endpoint, "items": api_data}
        )
        
        return {
            "record_id": record_id,
            "user_id": user_id,
            "endpoint": endpoint,
            "items_count": len(api_data),
            "message": "Data fetched and stored for user successfully"
        }
    except Exception as e:
        raise HTTPException(
            status_code=500,
            detail=f"Failed to fetch and store: {str(e)}"
        )

# Endpoint to get database statistics
@app.get("/db-stats")
async def get_database_stats():
    """
    Get statistics about stored data in the database.
    """
    with get_db_connection() as conn:
        cursor = conn.cursor()
        
        # Count cached API responses
        cursor.execute('SELECT COUNT(*) FROM api_responses')
        cached_count = cursor.fetchone()[0]
        
        # Count user records
        cursor.execute('SELECT COUNT(*) FROM fetched_data')
        records_count = cursor.fetchone()[0]
        
        # Get unique users
        cursor.execute('SELECT COUNT(DISTINCT user_id) FROM fetched_data')
        unique_users = cursor.fetchone()[0]
        
        # Get data types
        cursor.execute('SELECT DISTINCT data_type FROM fetched_data')
        data_types = [row[0] for row in cursor.fetchall()]
    
    return {
        "cached_api_responses": cached_count,
        "total_user_records": records_count,
        "unique_users": unique_users,
        "data_types": data_types,
        "database_file": DATABASE_FILE
    }

print("✓ API endpoints for database operations defined!")
````

## 11. Authentication & Authorization

Secure your API with JWT tokens, OAuth2, and role-based access control.

### Security Features:
- **JWT Tokens**: JSON Web Tokens for stateless authentication
- **Password Hashing**: Secure password storage
- **OAuth2 Integration**: Third-party authentication
- **Role-Based Access**: Different permissions for different users

````python
# Install: pip install python-jose cryptography passlib

from datetime import datetime, timedelta
from typing import Optional
import hashlib
import hmac

# ============================================================================
# JWT TOKEN AUTHENTICATION SETUP
# ============================================================================

SECRET_KEY = "your-secret-key-change-in-production"
ALGORITHM = "HS256"
ACCESS_TOKEN_EXPIRE_MINUTES = 30

# User database (in production, use real database)
fake_users_db = {
    "john": {"username": "john", "password": "hashed_password_123", "role": "admin"},
    "jane": {"username": "jane", "password": "hashed_password_456", "role": "user"}
}

class Token(BaseModel):
    access_token: str
    token_type: str

class TokenData(BaseModel):
    username: Optional[str] = None
    role: Optional[str] = None

class UserLogin(BaseModel):
    username: str
    password: str

# Simulate password hashing
def hash_password(password: str) -> str:
    """Hash password for storage"""
    return hashlib.sha256(password.encode()).hexdigest()

def verify_password(plain_password: str, hashed_password: str) -> bool:
    """Verify password"""
    return hash_password(plain_password) == hashed_password

def create_access_token(data: dict, expires_delta: Optional[timedelta] = None):
    """Create JWT token"""
    to_encode = data.copy()
    if expires_delta:
        expire = datetime.utcnow() + expires_delta
    else:
        expire = datetime.utcnow() + timedelta(minutes=15)
    
    to_encode.update({"exp": expire})
    # In production, use: jwt.encode(to_encode, SECRET_KEY, algorithm=ALGORITHM)
    return f"token_{to_encode.get('sub')}_simulated"

# Authentication endpoints
@app.post("/login", response_model=Token)
async def login(user: UserLogin):
    """
    Login endpoint to get JWT token.
    Returns access token for authenticated user.
    """
    if user.username not in fake_users_db:
        raise HTTPException(status_code=401, detail="Invalid credentials")
    
    db_user = fake_users_db[user.username]
    if not verify_password(user.password, db_user["password"]):
        raise HTTPException(status_code=401, detail="Invalid credentials")
    
    access_token_expires = timedelta(minutes=ACCESS_TOKEN_EXPIRE_MINUTES)
    access_token = create_access_token(
        data={"sub": user.username, "role": db_user["role"]},
        expires_delta=access_token_expires
    )
    
    return {
        "access_token": access_token,
        "token_type": "bearer"
    }

# Protected endpoint requiring authentication
@app.get("/protected")
async def protected_route(token: Optional[str] = None):
    """
    Protected endpoint that requires authentication token.
    """
    if not token:
        raise HTTPException(
            status_code=401,
            detail="Not authenticated",
            headers={"WWW-Authenticate": "Bearer"}
        )
    
    # In production, decode and verify the token
    return {"message": "Access granted", "token_type": "bearer"}

# Role-based access control
@app.get("/admin")
async def admin_only(role: Optional[str] = None):
    """
    Admin-only endpoint. Requires role='admin'.
    """
    if role != "admin":
        raise HTTPException(
            status_code=403,
            detail="Insufficient permissions"
        )
    
    return {"message": "Welcome admin!"}

print("✓ Authentication system configured!")
````

## 13. Dependency Injection

Use FastAPI's dependency injection system for code reuse and database management.

### Key Concepts:
- **Dependencies**: Reusable functions to inject into endpoints
- **Sub-dependencies**: Dependencies that depend on other dependencies
- **Database Sessions**: Manage database connections

````python
from fastapi import Depends

# ============================================================================
# DEPENDENCY INJECTION
# ============================================================================

# Simple dependency
async def get_query(q: Optional[str] = None):
    """Dependency for query parameter"""
    return {"q": q}

# Dependency with validation
async def validate_api_key(api_key: str = None):
    """Validate API key dependency"""
    if not api_key:
        raise HTTPException(status_code=401, detail="API key required")
    if api_key != "test-api-key":
        raise HTTPException(status_code=401, detail="Invalid API key")
    return api_key

# Sub-dependency (dependency that depends on another)
async def get_current_user(api_key: str = Depends(validate_api_key)):
    """Get current user based on API key"""
    return {"username": "test_user", "api_key": api_key}

# Database session dependency (simulated)
class DatabaseSession:
    def __init__(self):
        self.connection = "simulated_db_connection"
    
    def query(self, table):
        return f"Querying {table} from {self.connection}"

async def get_db():
    """Database session dependency"""
    db = DatabaseSession()
    try:
        yield db
    finally:
        print("Database connection closed")

# Using dependencies in endpoints
@app.get("/with-dependency")
async def endpoint_with_dependency(query: dict = Depends(get_query)):
    """Endpoint using dependency"""
    return {"received": query}

@app.get("/validate-api")
async def endpoint_with_validation(api_key: str = Depends(validate_api_key)):
    """Endpoint with validation dependency"""
    return {"api_key": api_key, "status": "valid"}

@app.get("/current-user")
async def get_user(user: dict = Depends(get_current_user)):
    """Get current user using sub-dependency"""
    return user

@app.get("/with-db")
async def endpoint_with_db(db: DatabaseSession = Depends(get_db)):
    """Endpoint with database dependency"""
    result = db.query("users")
    return {"database_query": result}

print("✓ Dependency injection configured!")
````

## 18. WebSockets

Enable real-time bidirectional communication with WebSockets.

### WebSocket Features:
- **Real-time Communication**: Send/receive data instantly
- **Connection Management**: Handle connections and disconnections
- **Broadcasting**: Send messages to multiple clients
- **Chat Applications**: Build collaborative features

````python
from fastapi import WebSocket, WebSocketDisconnect

# ============================================================================
# WEBSOCKET CONNECTIONS
# ============================================================================

# WebSocket connection manager
class ConnectionManager:
    def __init__(self):
        self.active_connections: List[WebSocket] = []
    
    async def connect(self, websocket: WebSocket):
        """Register new connection"""
        await websocket.accept()
        self.active_connections.append(websocket)
        print(f"Client connected. Total: {len(self.active_connections)}")
    
    def disconnect(self, websocket: WebSocket):
        """Remove disconnected client"""
        self.active_connections.remove(websocket)
        print(f"Client disconnected. Total: {len(self.active_connections)}")
    
    async def broadcast(self, message: str):
        """Broadcast message to all connected clients"""
        for connection in self.active_connections:
            try:
                await connection.send_text(message)
            except Exception as e:
                print(f"Error broadcasting: {e}")

manager = ConnectionManager()

# Simple WebSocket endpoint
@app.websocket("/ws")
async def websocket_endpoint(websocket: WebSocket):
    """
    Simple WebSocket endpoint.
    Test with: websocat ws://localhost:8000/ws
    """
    await manager.connect(websocket)
    try:
        while True:
            data = await websocket.receive_text()
            await websocket.send_text(f"Echo: {data}")
    except WebSocketDisconnect:
        manager.disconnect(websocket)
        await manager.broadcast(f"Clients: {len(manager.active_connections)}")

# Chat WebSocket endpoint
@app.websocket("/ws/chat/{client_id}")
async def websocket_chat(websocket: WebSocket, client_id: str):
    """
    Chat endpoint with client identification.
    """
    await manager.connect(websocket)
    await manager.broadcast(f"Client {client_id} joined the chat")
    
    try:
        while True:
            data = await websocket.receive_text()
            message = f"Client {client_id}: {data}"
            await manager.broadcast(message)
    except WebSocketDisconnect:
        manager.disconnect(websocket)
        await manager.broadcast(f"Client {client_id} left the chat")

# WebSocket with authentication (simulated)
@app.websocket("/ws/secure/{token}")
async def websocket_secure(websocket: WebSocket, token: str):
    """
    Secure WebSocket endpoint with token validation.
    """
    if token != "valid-token":
        await websocket.close(code=1008, reason="Invalid token")
        return
    
    await manager.connect(websocket)
    try:
        while True:
            data = await websocket.receive_text()
            await websocket.send_json({
                "message": data,
                "authenticated": True,
                "timestamp": datetime.now().isoformat()
            })
    except WebSocketDisconnect:
        manager.disconnect(websocket)

print("✓ WebSockets configured!")
````

````python
from functools import lru_cache
from fastapi.responses import Response

# ============================================================================
# PERFORMANCE OPTIMIZATION PATTERNS
# ============================================================================

# Response caching
cache = {}

@app.get("/cached-data/{item_id}")
async def get_cached_data(item_id: int):
    """
    Get data with caching.
    Subsequent requests return cached data.
    """
    if item_id in cache:
        print(f"Cache HIT for item {item_id}")
        return {"data": cache[item_id], "cached": True}
    
    print(f"Cache MISS for item {item_id}")
    # Simulate expensive operation
    await asyncio.sleep(1)
    data = {"id": item_id, "content": f"Data for item {item_id}"}
    cache[item_id] = data
    
    return {"data": data, "cached": False}

# Function-level caching with @lru_cache
@lru_cache(maxsize=32)
def expensive_computation(n: int) -> int:
    """Expensive computation with caching"""
    print(f"Computing for {n}...")
    return n ** 2

@app.get("/compute/{n}")
async def compute(n: int):
    """
    Compute with function-level caching.
    """
    result = expensive_computation(n)
    return {"input": n, "result": result}

# Response compression
@app.get("/large-response")
async def large_response():
    """
    Return large response.
    FastAPI automatically compresses with gzip.
    """
    data = [{"id": i, "value": f"Item {i}"} for i in range(1000)]
    return {"items": data, "count": len(data)}

# Optimized query-like endpoint
@app.get("/fast-search/{query}")
async def fast_search(query: str):
    """
    Optimized search using in-memory data structures.
    In production: use database indexes.
    """
    # Pre-built index (in memory)
    search_index = {
        "fastapi": ["FastAPI", "Python", "Web"],
        "python": ["Python", "Programming", "Backend"],
        "database": ["Database", "SQL", "Storage"]
    }
    
    result = search_index.get(query.lower(), [])
    
    return {
        "query": query,
        "results": result,
        "count": len(result)
    }

# Connection pooling simulation
class OptimizedDBConnection:
    def __init__(self, pool_size: int = 5):
        self.pool_size = pool_size
        self.active_connections = 0
    
    async def get_connection(self):
        """Get connection from pool"""
        if self.active_connections < self.pool_size:
            self.active_connections += 1
            print(f"Connection pooled: {self.active_connections}/{self.pool_size}")
            return {"connection": self.active_connections}
        return None
    
    async def release_connection(self):
        """Return connection to pool"""
        self.active_connections -= 1
        print(f"Connection released: {self.active_connections}/{self.pool_size}")

db_pool = OptimizedDBConnection(pool_size=10)

@app.get("/optimized-query")
async def optimized_query():
    """Use connection pool for query"""
    conn = await db_pool.get_connection()
    if conn:
        result = {"query_result": "data", "connection_id": conn["connection"]}
        await db_pool.release_connection()
        return result
    return {"error": "No available connections"}

print("✓ Performance optimizations configured!")
````

## 21. Rate Limiting & Throttling

Protect your API from abuse with rate limiting.

### Rate Limiting Features:
- **Request Limits**: Limit requests per client
- **User Quotas**: Different limits for different users
- **Time Windows**: Rolling or fixed time windows
- **DDoS Protection**: Prevent abuse

````python
# Install: pip install slowapi

from collections import defaultdict
from datetime import timedelta

# ============================================================================
# RATE LIMITING
# ============================================================================

# Simple rate limiting using in-memory store
class RateLimiter:
    def __init__(self, requests: int, window: int):
        """
        requests: number of requests allowed
        window: time window in seconds
        """
        self.requests = requests
        self.window = window
        self.clients = defaultdict(list)
    
    def is_allowed(self, client_id: str) -> bool:
        """Check if client is within rate limit"""
        now = time.time()
        
        # Remove old requests outside the window
        self.clients[client_id] = [
            req_time for req_time in self.clients[client_id]
            if now - req_time < self.window
        ]
        
        # Check if limit exceeded
        if len(self.clients[client_id]) >= self.requests:
            return False
        
        # Add current request
        self.clients[client_id].append(now)
        return True
    
    def get_remaining(self, client_id: str) -> int:
        """Get remaining requests for client"""
        return self.requests - len(self.clients[client_id])

# Create limiters with different rates
api_general_limiter = RateLimiter(requests=100, window=60)  # 100 per minute
api_search_limiter = RateLimiter(requests=10, window=60)    # 10 per minute
api_auth_limiter = RateLimiter(requests=5, window=300)      # 5 per 5 minutes

# Rate limiting middleware
@app.middleware("http")
async def rate_limit_middleware(request, call_next):
    """Check rate limits on all requests"""
    client_ip = request.client.host if request.client else "unknown"
    
    # Skip rate limiting for certain paths
    if request.url.path in ["/health", "/"]:
        return await call_next(request)
    
    # Apply general rate limit
    if not api_general_limiter.is_allowed(client_ip):
        raise HTTPException(
            status_code=429,
            detail="Too many requests. Try again later."
        )
    
    response = await call_next(request)
    response.headers["X-RateLimit-Remaining"] = str(
        api_general_limiter.get_remaining(client_ip)
    )
    
    return response

# Per-endpoint rate limiting
@app.get("/search-limited")
async def limited_search(q: str):
    """
    Search endpoint with stricter rate limiting.
    """
    client_ip = "default_client"  # In production, get from request
    
    if not api_search_limiter.is_allowed(client_ip):
        raise HTTPException(
            status_code=429,
            detail="Search API rate limit exceeded. Max 10 requests per minute."
        )
    
    return {
        "query": q,
        "results": ["result1", "result2"],
        "remaining": api_search_limiter.get_remaining(client_ip)
    }

# Authentication rate limiting
@app.post("/login-limited")
async def login_limited(credentials: UserLogin):
    """
    Login endpoint with rate limiting to prevent brute force.
    """
    client_ip = "default_client"
    
    if not api_auth_limiter.is_allowed(client_ip):
        raise HTTPException(
            status_code=429,
            detail="Too many login attempts. Try again in 5 minutes."
        )
    
    # Simulate authentication
    return {"access_token": "token_123", "token_type": "bearer"}

# User-based quota system
class UserQuota:
    def __init__(self):
        self.quotas = {
            "free": 1000,      # 1000 requests
            "premium": 10000,  # 10000 requests
            "enterprise": None # unlimited
        }
        self.usage = defaultdict(int)
    
    def check_quota(self, user_id: str, plan: str) -> bool:
        """Check if user is within quota"""
        quota = self.quotas.get(plan)
        
        if quota is None:
            return True  # Unlimited
        
        if self.usage[user_id] >= quota:
            return False
        
        self.usage[user_id] += 1
        return True
    
    def get_usage(self, user_id: str, plan: str) -> dict:
        """Get usage information"""
        quota = self.quotas.get(plan)
        usage = self.usage[user_id]
        
        if quota is None:
            return {"quota": "unlimited", "usage": usage}
        
        return {
            "quota": quota,
            "usage": usage,
            "remaining": quota - usage,
            "percentage": (usage / quota) * 100
        }

quota_manager = UserQuota()

@app.get("/api/data/{user_id}")
async def get_user_data(user_id: str, plan: str = "free"):
    """
    API endpoint with usage-based quotas.
    """
    if not quota_manager.check_quota(user_id, plan):
        usage = quota_manager.get_usage(user_id, plan)
        raise HTTPException(
            status_code=429,
            detail=f"Quota exceeded for {plan} plan. Usage: {usage['usage']}/{usage['quota']}"
        )
    
    usage = quota_manager.get_usage(user_id, plan)
    
    return {
        "data": "Your data",
        "quota_usage": usage
    }

# Rate limit status endpoint
@app.get("/rate-limit-status")
async def rate_limit_status():
    """Get current rate limiting status"""
    return {
        "general_limit": {
            "requests": api_general_limiter.requests,
            "window_seconds": api_general_limiter.window
        },
        "search_limit": {
            "requests": api_search_limiter.requests,
            "window_seconds": api_search_limiter.window
        },
        "auth_limit": {
            "requests": api_auth_limiter.requests,
            "window_seconds": api_auth_limiter.window
        }
    }

print("✓ Rate limiting and throttling configured!")
````
